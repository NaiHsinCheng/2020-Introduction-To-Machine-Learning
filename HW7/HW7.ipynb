```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
"""第一題"""
#創建函數
x_data = np.linspace(-1, 1, 100)    #x_data有100列 一個特性100個例子
y_data = x_data**3 + 2*x_data**2 - 3*x_data - 1
```

```python
#定義模型
model3_10 = tf.keras.Sequential([        #三層、十個神經元
    tf.keras.layers.Dense(10,input_shape=(1,),activation="relu"),
    tf.keras.layers.Dense(1)
])
model3_100 = tf.keras.Sequential([        #三層、一百個神經元
    tf.keras.layers.Dense(100,input_shape=(1,),activation="relu"),
    tf.keras.layers.Dense(1)
])
model3_1000 = tf.keras.Sequential([        #三層、一千個神經元
    tf.keras.layers.Dense(1000,input_shape=(1,),activation="relu"),
    tf.keras.layers.Dense(1)
])
model3_10000 = tf.keras.Sequential([        #三層、一萬個神經元
    tf.keras.layers.Dense(10000,input_shape=(1,),activation="relu"),
    tf.keras.layers.Dense(1)
])
model5_10 = tf.keras.Sequential([        #五層、十個神經元
    tf.keras.layers.Dense(10,input_shape=(1,),activation="relu"),
    tf.keras.layers.Dense(10,input_shape=(1,),activation="relu"),
    tf.keras.layers.Dense(10,input_shape=(1,),activation="relu"),
    tf.keras.layers.Dense(1)
])
model5_100 = tf.keras.Sequential([        #五層、一百個神經元
    tf.keras.layers.Dense(100,input_shape=(1,),activation="relu"),
    tf.keras.layers.Dense(100,activation="relu"),
    tf.keras.layers.Dense(100,activation="relu"),
    tf.keras.layers.Dense(1)
])

#設置訓練參數
model3_10.compile(optimizer="adam", loss="mse")
model3_100.compile(optimizer="adam", loss="mse")
model3_1000.compile(optimizer="adam", loss="mse")
model3_10000.compile(optimizer="adam", loss="mse")
model5_10.compile(optimizer="adam", loss="mse")
model5_100.compile(optimizer="adam", loss="mse")


#訓練
history3_10 = model3_10.fit(x_data, y_data, epochs=100)
history3_100 = model3_100.fit(x_data, y_data, epochs=100)
history3_1000 = model3_1000.fit(x_data, y_data, epochs=100)
history3_10000 = model3_10000.fit(x_data, y_data, epochs=100)
history5_10 = model5_10.fit(x_data, y_data, epochs=100)
history5_100 = model5_100.fit(x_data, y_data, epochs=100)


#取得結果
y_predict3_10 = model3_10.predict(x_data)
y_predict3_100 = model3_100.predict(x_data)
y_predict3_1000 = model3_1000.predict(x_data)
y_predict3_10000 = model3_10000.predict(x_data)
y_predict5_10 = model5_10.predict(x_data)
y_predict5_100 = model5_100.predict(x_data)
```

```python
#作圖
plt.subplot(2,4,1)
plt.plot(x_data, y_data)
plt.plot(x_data, y_predict3_10)
plt.title("3/10")

plt.subplot(2,4,2)
plt.plot(x_data, y_data)
plt.plot(x_data, y_predict3_100)
plt.title("3/100")

plt.subplot(2,4,3)
plt.plot(x_data, y_data)
plt.plot(x_data, y_predict3_1000)
plt.title("3/1000")

plt.subplot(2,4,4)
plt.plot(x_data, y_data)
plt.plot(x_data, y_predict3_10000)
plt.title("3/10000")

plt.subplot(2,4,5)
plt.plot(x_data, y_data)
plt.plot(x_data, y_predict5_10)
plt.title("5/10")

plt.subplot(2,4,6)
plt.plot(x_data, y_data)
plt.plot(x_data, y_predict5_100)
plt.title("5/100")

plt.show()
```

```python
"""第二題"""
#創建資料
circles_data, circles_data_labels = datasets.make_circles(n_samples=50, factor=0.1, noise=0.1)

for j in range(50):
    if circles_data_labels[j] ==0:
        plt.scatter(circles_data[j,0],circles_data[j,1],color="r")
    else:
        plt.scatter(circles_data[j,0],circles_data[j,1],color="b")
plt.show()
```

```python
#建立類神經網路模型
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100,input_shape=(2,),activation="relu"),
    tf.keras.layers.Dense(100,activation="relu"),
    tf.keras.layers.Dense(100,activation="softmax"),
    tf.keras.layers.Dense(1)
])

#設置訓練參數
model.compile(optimizer="adam", loss="mse")
```

```python
# 使用類神經網路
history =  model.fit(circles_data, circles_data_labels, epochs=100)
# 接下來進行視覺化, 要想進行視覺化, 我們核心就是要呼叫plt.contour函式畫圖, 但是它要求傳入三個矩陣, 
#而我們的x1和x2為向量, 預測的值也為向量, 所有我們需要將x1和x2轉換為矩陣

# 獲取邊界範圍, 為了產生資料
x1_min, x1_max = np.min(circles_data[:, 0]) - 0.5, np.max(circles_data[:, 0]) + 0.5
x2_min, x2_max = np.min(circles_data[:, 1]) - 0.5, np.max(circles_data[:, 1]) + 0.5

# 生成新的資料, 並呼叫meshgrid網格搜尋函式幫助我們生成矩陣
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.2), np.arange(x2_min, x2_max, 0.2))
# 有了新的資料, 我們需要將這些資料輸入到分類器獲取到結果, 但是因為輸入的是矩陣, 我們需要給你將其轉換為符合條件的資料
Z = model.predict(np.c_[xx1.ravel(), xx2.ravel()])
```

```python
Z=[round(i) for i in Z[:,0]]
# 這個時候得到的是Z還是一個向量, 將這個向量轉為矩陣即可
Z = np.reshape(Z,xx1.shape)
```

```python
from matplotlib.colors import ListedColormap
plt.figure()
# 為什麼需要輸入矩陣, 因為等高線函式其實是3D函式, 3D座標是三個平面, 平面對應矩陣
plt.contourf(xx1, xx2, Z, 1,cmap=ListedColormap(['g','r']),alpha=0.2)
for j in range(50):
    if circles_data_labels[j] ==0:
        plt.scatter(circles_data[j,0],circles_data[j,1],color="g")
    else:
        plt.scatter(circles_data[j,0],circles_data[j,1],color="r")
plt.show()

```
