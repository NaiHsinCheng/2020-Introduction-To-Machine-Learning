```python
"""
   Author：B062040027 鄭乃心 / B062040020 張詠晴
   Date：2020.09.23
   Purpose：機器學習導論Homework#2
"""
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt

#讀資料
with open('HW2_data.csv',newline='') as csvfile:
    rows = csv.reader(csvfile)
    name=list()
    for row in rows:
        check=1
        for n in name:
            if row[0]==n:
                check=0
        if check:
            name.append(row[0])
    del name[0]

    temp=list()
    for i in name:
        inside=list([' ',' ',' ',' ',' ',' '])
        temp.append(inside)
    dictionary=dict(zip(name,temp))
    
with open('HW2_data.csv',newline='') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        if row[1]=='Sex':
            dictionary[row[0]][0]=row[2]
        elif row[1]=='Age':
            dictionary[row[0]][1]=row[2]
        elif row[1]=='HR':
            dictionary[row[0]][2]=row[2]
        elif row[1]=='Height':
            dictionary[row[0]][3]=row[2]
        elif row[1]=='Weight':
            dictionary[row[0]][4]=row[2]
        elif row[1]=='BP':
            dictionary[row[0]][5]=row[2]
            
#建立資料 計算個值平均 找出各值feature
sex=list()              #建立要做成dataframe的資料
age=list()
hr=list()
height=list()
weight=list()
bp=list()
ageAverage=0;ageCount=0           #平均
hrAverage=0;hrCount=0
heightAverage=0;heightCount=0
weightAverage=0;weightCount=0
bpAverage=0;bpCount=0
ageFeature=0           #feature
hrFeature=0
heightFeature=0
weightFeature=0
bpFeature=0

for n in name:
    sex.append(dictionary[n][0])
    age.append(dictionary[n][1])
    if dictionary[n][1] != ' ':
        ageAverage+=int(dictionary[n][1])
        ageCount+=1
        if int(dictionary[n][1])>ageFeature:
            ageFeature=int(dictionary[n][1])
            
    hr.append(dictionary[n][2])
    if dictionary[n][2] != ' ':
       hrAverage+=int(dictionary[n][2])
       hrCount+=1
       if int(dictionary[n][2])>hrFeature:
           hrFeature=int(dictionary[n][2])
        
    height.append(dictionary[n][3])
    if dictionary[n][3] != ' ':
        heightAverage+=int(dictionary[n][3])
        heightCount+=1
        if int(dictionary[n][3])>heightFeature:
            heightFeature=int(dictionary[n][3])
            
    weight.append(dictionary[n][4])
    if dictionary[n][4] != ' ':
        weightAverage+=int(dictionary[n][4])
        weightCount+=1
        if int(dictionary[n][4])>weightFeature:
            weightFeature=int(dictionary[n][4])
        
    bp.append(dictionary[n][5])
    if dictionary[n][5] != ' ':
        bpAverage+=int(dictionary[n][5])
        bpCount+=1
        if int(dictionary[n][5])>bpFeature:
           bpFeature=int(dictionary[n][5])

ageAverage/=ageCount
hrAverage/=hrCount
heightAverage/=heightCount
weightAverage/=weightCount
bpAverage/=bpCount

#統計男女人數(圓餅圖) 男生身高體重(散部圖) 女生身高體重(散部圖)
men=0;women=0
FH=list();FW=list()
MH=list();MW=list()
for n in name:
    if dictionary[n][0] == 'F':
        women+=1
        if dictionary[n][3] != ' ':
            FH.append(dictionary[n][3])
        else:
            FH.append(int(heightAverage))
        if dictionary[n][4] != ' ':
            FW.append(dictionary[n][4])
        else:
            FW.append(int(weightAverage))
    elif dictionary[n][0] == 'M':
        men+=1
        if dictionary[n][3] != ' ':
            MH.append(dictionary[n][3])
        else:
            MH.append(int(heightAverage))
        if dictionary[n][4] != ' ':
            MW.append(dictionary[n][4])
        else:
            MW.append(int(weightAverage))

#統整年齡區間
a=[0,0,0,0,0,0,0,0]       #年齡區間的list
for n in name:
    if dictionary[n][1] != ' ':
        if int(dictionary[n][1])<=10:
            a[0]+=1
        elif int(dictionary[n][1])<=20:
            a[1]+=1
        elif int(dictionary[n][1])<=30:
            a[2]+=1
        elif int(dictionary[n][1])<=40:
            a[3]+=1
        elif int(dictionary[n][1])<=50:
            a[4]+=1
        elif int(dictionary[n][1])<=60:
            a[5]+=1
        elif int(dictionary[n][1])<=70:
            a[6]+=1
        elif int(dictionary[n][1])<=80:
            a[7]+=1
            
###繪圖###

#身高體重散布圖
plt.plot(FW,FH,'r.')
plt.plot(MW,MH,'b.')
plt.ylabel('Height')
plt.xlabel('Weight')
plt.title("Height/Weight")
plt.show()

#直方圖
X1=['1-10','11-20','21-30','31-40','41-50','51-60','61-70','71-80',]
Y1=[a[0],a[1],a[2],a[3],a[4],a[5],a[6],a[7]]
plt.ylabel('count')
plt.xlabel('Age')
plt.bar(X1,Y1)
plt.title('Age')
plt.show() 

#圓餅圖
lable=['1-10','11-20','21-30','31-40','41-50','51-60','61-70','71-80']
values=[a[0],a[1],a[2],a[3],a[4],a[5],a[6],a[7]]
#color=['red','blue']
plt.pie(values,labels=lable)
plt.title('Age')
plt.axis('equal')
plt.show()              

#繪製男女性別比圓餅圖
lable=['Female','Male']
values=[women,men]
color=['red','blue']
plt.pie(values,labels=lable,colors=color)
plt.title("Female/Male")
plt.axis('equal')
plt.show()

#dataframe
dictionary={'Name':name,'Sex':sex,'Age':age,'HR':hr,'Height':height,'Weight':weight,'Bp':bp}
df=pd.DataFrame(dictionary)
df=df.append({'Name':'Average','Sex':' ','Age':ageAverage,'HR':hrAverage,'Height':heightAverage,'Weight':weightAverage,'Bp':bpAverage},ignore_index=True)

print('Feature')
print('Age',ageFeature)
print('HR',hrFeature)
print('Height',heightFeature)
print('Weight',weightFeature)
print('BP',bpFeature)

df
```
